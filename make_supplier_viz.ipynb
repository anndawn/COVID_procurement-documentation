```python
import pandas as pd
import re
```

```python
df_supplier=pd.read_csv("data//data_original//supplier_data.csv")
df=pd.read_csv("data//data_original//1113_data.csv")
df_institution=pd.read_csv("data//item_department_match//institution_match.csv")
```

```python
df.loc[df["Institution_name"]=="Education.","Institution_name"]="Gauteng - education"
```

```python
df["Institution_name"]=df["Institution_name"].apply(lambda x:str(x).upper())
df["Institution_name"]=df["Institution_name"].apply(lambda x:re.sub(r"\.","",x))
df_institution["Original_name"]=df_institution["Original_name"].apply(lambda x:x.upper())
```

```python
df_use=pd.merge(df,df_institution,how="left",left_on="Institution_name",right_on="Original_name") 
```

```python
df_use.drop(columns=["Original_name",' filter'],inplace=True)
```

```python
df_supplier_clean=df_supplier.drop_duplicates(subset="CSD_Registration_No",keep="first")
```

```python
def clean_str(string):
    string=string.strip().lower()
    string=re.sub(r"\.","",string)
    string=re.sub("\s+"," ",string)
    
    return string
```

```python
df_supplier_clean.loc[:,"Supplier_name"]=df_supplier_clean.loc[:,"Supplier_name"].apply(lambda x: clean_str(x))
df_supplier_clean.loc[:,"CSD_Registration_No"]=df_supplier_clean.loc[:,"CSD_Registration_No"].apply(lambda x: clean_str(x))
```

```python
df_supplier_clean.loc[df_supplier_clean["Director"].isna(),"Director"]=" "
```

```python
Supplier_analysis=df_supplier_clean.groupby(by=['CSD_Registration_No','Director'])["pay_num"].sum().to_frame()
Supplier_analysis.reset_index(inplace=True)
Supplier_analysis.sort_values(by="pay_num",ascending=False,inplace=True)
```

```python
Supplier_analysis.head(60).to_csv("data//data_manual//supplier_analysis.csv",index=False)
```

```python
df_analysis=pd.read_csv("data//data_manual//supplier_analysis_cleaned.csv")
```

```python
df_analysis["supplier_name"]=df_analysis["supplier_name"].apply(lambda x: x.split("\n"))
df_analysis["CSD_Registration_No"]=df_analysis["CSD_Registration_No"].apply(lambda x: x.upper())
```

```python
df_analysis
```

```python
df_analysis=pd.merge(df_analysis,df_supplier.loc[:,["CSD_Registration_No","Director"]],how="left",on=["CSD_Registration_No","Director"])
df_analysis
```

```python
df_analysis_flat=df_analysis.loc[:,["CSD_Registration_No","pay_num","supplier_name","Director"]].explode("supplier_name")
```

```python
df_analysis_flat["supplier_name"]=df_analysis_flat["supplier_name"].apply(lambda x: clean_str(x))
df_use["Supplier_name"]=df_use["Supplier_name"].apply(lambda x: clean_str(str(x)))
```

```python
df_analysis_flat.drop_duplicates(inplace=True)
df_analysis_flat.reset_index(inplace=True,drop=True)
```

```python
df_orders=pd.merge(df_analysis_flat,df_use,how="left",left_on="supplier_name",right_on="Supplier_name")
```

```python
df_orders.rename(columns={"pay_num_x":"supplier_sum","pay_num_y":"pay_num"},inplace=True)
```

```python
# filter out supplier name that doesn't belong to the supplier, but might appeared in the drop down list becasue of a wrong entry
filter1057=((df_orders["supplier_name"]=="masiqhame trading")&(df_orders["Institution_name"]=="ENGCOBO LOCAL MUNICIPALITY"))
filternlhs=((df_orders["supplier_name"]=="national health laboratory services")&(df_orders["Institution_name"]=="GAUTENG DEPARTMENT OF HEALTH"))
filter_medplant=((df_orders["supplier_name"]=="medical plant africa")&(df_orders["pay_num"]==241500))
```

```python
df_orders_filter=df_orders.loc[(~filter_medplant)&(~filternlhs)&(~filter1057),:]
```

```python
df_orders_filter.drop(columns=["Supplier_name"],inplace=True)
df_orders_filter.reset_index(inplace=True,drop=True)
```

```python
# df_analysis_flat["Director"]=df_analysis_flat["Director"].apply(lambda x:x.strip())
supplier_flat=pd.merge(df_supplier_clean,df_analysis_flat,how="right",on=["Director","pay_num"])
supplier_flat.drop(columns=["CSD_Registration_No_x","Count_of_institutions_serviced","No_of_orders"],inplace=True)
supplier_flat.rename(columns={"pay_num":"pay_num_dashboard","CSD_Registration_No_y":"CSD_Registration_No","Supplier_name":"Cleaned_supplier"},inplace=True)
supplier_flat.reset_index(inplace=True,drop=True)
supplier_flat
```

```python
df_order_supplier=pd.merge(df_orders_filter,supplier_flat,how="left",on="supplier_name")
df_order_supplier.drop(columns=["CSD_Registration_No_x","Director_x","order_num_y","Director_y","pay_num_dashboard"],inplace=True)
df_order_supplier.rename(columns={"CSD_Registration_No_y":"CSD_Registration_No","order_num_x":"order_num","Supplier_name":"Cleaned_supplier"},inplace=True)
df_order_supplier
```

```python
No_orders=df_order_supplier.groupby(by=["CSD_Registration_No","Cleaned_supplier"])["Cleaned_name"].count().to_frame()
No_orders.rename(columns={"Cleaned_name":"No_orders"},inplace=True)
No_orders.reset_index(inplace=True)
```

```python
No_institutions=df_order_supplier.groupby(by=["CSD_Registration_No","Cleaned_supplier"])["Cleaned_name"].nunique().to_frame()
No_institutions.rename(columns={"Cleaned_name":"No_Institutions_served"},inplace=True)
No_institutions.reset_index(inplace=True)
```

```python
amount_agg=df_order_supplier.groupby(by=["CSD_Registration_No","Cleaned_supplier"]).agg({'pay_num':'sum','order_num':'sum'})
amount_agg.rename(columns={"pay_num":"total_payment","order_num":"total_order"},inplace=True)
amount_agg.reset_index(inplace=True)
```

```python
from functools import reduce
dfs = [No_orders, No_institutions,amount_agg]
df_viz = reduce(lambda left,right: pd.merge(left,right,on=['CSD_Registration_No','Cleaned_supplier']), dfs)
```

```python
df_viz.sort_values(by="total_payment",ascending=False,inplace=True)
```

```python
df_viz.loc[df_viz["No_Institutions_served"]<3,"Institution category"]="Only served 1-2 institutions"
df_viz.loc[(df_viz["No_Institutions_served"]<10)&(df_viz["No_Institutions_served"]>2),"Institution category"]="Serves 3-9 institutions"
df_viz.loc[df_viz["No_Institutions_served"]>9,"Institution category"]="Serves at least 10 institutions"
```

```python
df_viz.to_csv("data//data_viz//supplier_summary.csv")
```
